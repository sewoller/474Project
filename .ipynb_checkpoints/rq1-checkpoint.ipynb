```python
#import libraries
import pandas as pd
from matplotlib import pyplot as plt

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
# load datasets
a = pd.read_csv('https://media.githubusercontent.com/media/sewoller/474Project/main/Answers_Processed.csv', low_memory=False)
q = pd.read_csv('https://media.githubusercontent.com/media/sewoller/474Project/main/Questions_Processed.csv', low_memory=False)
```

### RQ1: What programming languages have the highest/lowest rated questions on stack overflow? 

We can begin our investigation by visualizing the distribution of scores for each language. 

```python
print("Questions: Score by Language")
q["Score"].hist(by=q["Tag"], bins = 50)
```

The distributions have such a large number of scores set to zero that the histograms are not much help, we cannot see much detail. We shall look at the mean, max, and min for the scores to get a better idea of the distributions. 

```python
questions_score_stats = q.groupby('Tag').agg({'Score': ['mean', 'min', 'max', 'count']})
questions_score_stats
```

```python
df = q.groupby(['Tag'])['Score'].max()
df.plot(kind='bar', title='Max Score', ylabel='Max Score', xlabel='Language')
```

We can see that the language with the highest scored question is java, with a score of 3613. This may not be significant, because java is also the language with the highest number of questions. Therefore it may be scored higher simply because a larger amount of people are interested in java questions. We could test this by dividing the maximum score by the number of questions to get a normalized value for each tag. 

```python
df1 = q.value_counts('Tag')
df2 = q.groupby(['Tag'])['Score'].max()
df = pd.merge(df1, df2, how='inner', on=['Tag'])
df['normal_max'] = df['Score'] / df['count']
df.plot( y = 'normal_max', kind='bar', title='Normalized Max Score', ylabel='Mean Score', xlabel='Language')
```

After normalization, java still has a significantly high score. HTML becomes the highest score holder after normalization, while it was the seventh biggest max score before normalization.

We can create a scatterplot to visualize the correlation between language popularity and maximum question score. 

```python
df1 = q.value_counts('Tag')
df2 = q.groupby(['Tag'])['Score'].max()
df = pd.merge(df1, df2, how='inner', on=['Tag'])
df.plot(x = 'count', y = 'Score', kind='scatter')
```

We can see a clear positive correlation between the number of questions that a language has and the maximum score of the questions. Therefore, the maximum score will tell us more about the popularity of the language than the quality of the questions. 

```python
df = q.groupby(['Tag'])['Score'].mean()
df.plot(kind='bar', title='Mean Score', ylabel='Mean Score', xlabel='Language')
```

The language with the highest mean score is python. The average score for python 6.44, which is significantly higher than the second highest mean, which is ruby at 5.3. 

```python
df1 = q.value_counts('Tag')
df2 = q.groupby(['Tag'])['Score'].mean()
df = pd.merge(df1, df2, how='inner', on=['Tag'])
df.plot(x = 'count', y = 'Score', kind='scatter')
```

The above scatterplot shows each language's mean question score against the total number of questions. There is no visually evident correlation between these values, which suggests that different languages may indeed being in a different quality of questions. We can see, for example, that while php has a high number of questions, it has a significantly low mean score for those questions.
